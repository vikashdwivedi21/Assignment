```python
"""* **Bar plot:** A bar plot is a graphical representation of the distribution of data.
It is used to compare the values of different categories or groups.

* **Histogram:** A histogram is a graphical representation of the distribution of data. It is used to visualize
the shape of the data and to identify the presence of outliers.

* **Scatter plot:** A scatter plot is a graphical representation of the relationship between two variables.
It is used to identify the correlation between the variables and to identify outliers.

* **Line plot:** A line plot is a graphical representation of the change in a variable over time.
It is used to track the trend of the variable and to identify patterns.

* **Box plot:** A box plot is a graphical representation of the distribution of data. 
It is used to identify the median, quartiles, and outliers"""
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

sns.set()
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
plt.title("Heatmap Flight Data")
plt.show()
```
