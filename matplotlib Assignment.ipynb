```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot

"""

-->A bar plot, also known as a bar chart, is a type of chart that presents categorical data with rectangular bars. Each bar represents a category, and the length or height of the bar corresponds to the value or frequency of that category. Bar plots are commonly used to visualize and compare different categories or groups.

Here are some key reasons why bar plots are used:

1. **Comparison of categorical data**: Bar plots are effective for comparing and visualizing categorical data. They provide a clear and intuitive way to compare the values or frequencies of different categories, making it easy to identify patterns, trends, and differences between groups.

2. **Data exploration and analysis**: Bar plots help in exploring and analyzing data. They can reveal insights about the distribution of categorical variables, highlight dominant or minority categories, and identify any outliers or unusual patterns.

3. **Presentation of summaries and summaries**: Bar plots are useful for presenting summary statistics or aggregated data. They can display means, counts, percentages, or any other summary measure for each category, making it easier to communicate key findings or results.

4. **Visualizing survey results and poll data**: Bar plots are commonly used to present survey results or poll data, where categories represent different response options. They provide a straightforward way to display the distribution of responses and show the relative popularity or frequency of each option.

5. **Comparison of multiple groups**: Bar plots can be used to compare multiple groups or subcategories within each main category. By using grouped or stacked bar plots, you can display the distribution or comparison of different groups within each category, providing a comprehensive view of the data.

6. **Track changes over time**: Bar plots can also be used to track changes in categorical data over time. By plotting the same categories on a time axis, you can visualize how the values or frequencies of the categories evolve, allowing for trend analysis and identifying any shifts or patterns.

Overall, bar plots are a versatile and widely used visualization tool for categorical data. They provide an effective way to summarize, compare, and communicate information about different categories or groups, making them valuable in various fields such as statistics, data analysis, market research, and social sciences."""
```

```python
#Q5.What is a box plot? Why is it used? Using the following data plot a box plot.


"""-->Box plot, is a graphical representation of the distribution of a dataset. It displays a summary of the data's central tendency, spread, and any potential outliers. The plot consists of a rectangular box (the "box") and two lines (the "whiskers") that extend from the box. Box plots are commonly used to visually analyze and compare the distribution of numerical data across different categories or groups.

Here's how a box plot is constructed and why it is used:

1. **Central tendency**: The box in a box plot represents the interquartile range (IQR), which encompasses the middle 50% of the data. The horizontal line inside the box represents the median, which indicates the central tendency or the midpoint of the data.

2. **Spread**: The whiskers in a box plot represent the variability or spread of the data. Typically, they extend from the box to the maximum and minimum values within a certain range. However, the exact definition of the whiskers can vary depending on the specific implementation or conventions.

3. **Outliers**: Box plots also provide a visual indication of potential outliers in the data. Outliers are data points that significantly deviate from the rest of the distribution. They are displayed as individual points or asterisks outside the whiskers, indicating values that are considered unusually high or low.

By using box plots, you can achieve the following:

- **Identify distribution characteristics**: Box plots help you understand the shape and spread of a dataset. You can quickly observe whether the data is symmetrically distributed or skewed, as well as the presence of outliers or extreme values.

- **Compare distributions**: Box plots are particularly useful for comparing the distributions of multiple datasets or different groups. By placing multiple box plots side by side, you can easily compare their medians, spreads, and skewness. This allows you to analyze the similarities, differences, or variations between the groups.

- **Detect outliers**: Box plots provide a clear visual representation of potential outliers in the data. Outliers can be identified as individual data points outside the whiskers, giving you an indication of extreme values that may require further investigation or consideration.

- **Summarize data**: Box plots provide a concise summary of the central tendency, spread, and outliers in a dataset. They offer a compact representation of key statistics, such as the median, quartiles, and range, allowing for quick and effective communication of data characteristics.

Box plots are widely used in various fields, including statistics, data analysis, exploratory data analysis, and inferential statistics. They offer a valuable tool for visualizing and summarizing numerical data, aiding in understanding distributions, making comparisons, and identifying outliers or extreme values."""
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Create a box plot for box1
axs[0].boxplot(box1)
axs[0].set_title('Box 1')
axs[0].set_ylabel('Values')

# Create a box plot for box2
axs[1].boxplot(box2)
axs[1].set_title('Box 2')
axs[1].set_ylabel('Values')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()







```
