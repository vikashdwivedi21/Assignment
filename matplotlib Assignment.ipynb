```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Create a box plot for box1
axs[0].boxplot(box1)
axs[0].set_title('Box 1')
axs[0].set_ylabel('Values')

# Create a box plot for box2
axs[1].boxplot(box2)
axs[1].set_title('Box 2')
axs[1].set_ylabel('Values')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
In this code, we create a figure with two subplots using plt.subplots(1, 2, figsize=(10, 4)). This will generate two side-by-side subplots. The figsize parameter is used to set the size of the figure.

We then create a box plot for box1 on the first subplot (axs[0]) using axs[0].boxplot(box1). Similarly, we create a box plot for box2 on the second subplot (axs[1]) using axs[1].boxplot(box2).

To set the titles and y-axis labels for each subplot, we use axs[0].set_title(), axs[0].set_ylabel() for the first subplot, and axs[1].set_title(), axs[1].set_ylabel() for the second subplot.

We use plt.tight_layout() to adjust the spacing between the subplots and ensure they are properly arranged.

Finally, we display the plot using plt.show().

Running this code will generate a figure with two box plots side by side, each representing the distributions of box1 and box2 separately.






```
